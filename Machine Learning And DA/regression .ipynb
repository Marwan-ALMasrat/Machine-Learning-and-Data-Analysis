```python
def percentage_with_question_mark(column):
    # Convert the column to string type before using .str accessor
    count_question_marks = column.astype(str).str.contains('\?').sum()
    total_values = len(column)
    return (count_question_marks / total_values) * 100

percentage_values = df.apply(percentage_with_question_mark )

print(percentage_values)
```

*  All of the missing values ​​filled with "?" in each column are very few,
except (normalized-losses)column has 20%.
 * Beacause we have a little of values represent "?" in some rows I'm gonna drop it, in (normalized-losses) column I will use medain to fill these values.


* Some of column the data type should be a float or int but it show as object.

 * So we have to change them to numeric:

    * normalized-losses, stroke, horesepower, peak-rpm, price(object) -> numeric.

```python
# Replace '?' with NaN to represent missing values
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')

df['normalized-losses'].fillna(df['normalized-losses'].median(), inplace=True)
```

```python
#drop every column "?"
df= df[~df.isin(['?']).any(axis=1)]
```

```python
df['stroke'] = pd.to_numeric(df['stroke'], errors='coerce')
df['bore'] = pd.to_numeric(df['bore'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

# Machine Learning
  ## Train | Test Split, Scaling and Dealing with the categorical Data.

```python
df.info()
```

* Convert the categorical columns to numeric data by Binary Encoding.
* Create a StandardScaler object and normalize the X train and test set feature data.



```python
!pip install category_encoders
import category_encoders as ce
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop('price', axis=1)
y = df['price']

binary_cols = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style',
               'drive-wheels', 'engine-location', 'engine-type',
               'num-of-cylinders', 'fuel-system']
encoder = ce.BinaryEncoder(cols=binary_cols)

X = encoder.fit_transform(X)

standard_cols = ['symboling', 'normalized-losses','wheel-base', 'length', 'width', 'height',
                 'curb-weight', 'engine-size', 'bore', 'stroke',
                 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']
scaler = StandardScaler()

X[standard_cols] = scaler.fit_transform(X[standard_cols])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)


```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error # Import mean_absolute_error
import numpy as np # Import numpy for np.sqrt

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)

print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
print("R2:",r2)
```

# Select The Best Model


* Feature Importances

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

importances = rf.feature_importances_


sorted_idx = np.argsort(importances)[::-1]


print("Feature importances (sorted):")
for idx in sorted_idx:
    print(f"Feature {X_train.columns[idx]}: Importance {importances[idx]}")


plt.figure(figsize=(13, 8))
plt.title("Feature Importance from Random Forest")
plt.barh(range(X_train.shape[1]), importances[sorted_idx], align="center")
plt.yticks(range(X_train.shape[1]), [f"{X_train.columns[i]}" for i in sorted_idx])

plt.xlabel("Importance")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score



models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, random_state=42)
}

results = {}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)


    results[model_name] = {
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R² Score": r2
    }

results_df = pd.DataFrame(results).T

print(results_df)

best_model = results_df["R² Score"].idxmax()
print(f"\nThe best model based on R² Score is: {best_model}")
```
