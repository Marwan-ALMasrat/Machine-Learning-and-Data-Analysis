### Before filling in a missing values:


```python
df_num = df.select_dtypes(include=["float64"])

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

# Number of columns to create subplots
num_columns = len(df_num.columns)

# Create a figure with subplots (1 row, 2 columns for each column in df_num)
fig, axes = plt.subplots(num_columns, 2, figsize=(12, 6 * num_columns))

for i, column in enumerate(df_num.columns):
    # Get the data for the current column
    data = df_num[column].dropna()

    # Plot the histogram and the normal distribution fit (on the first axis)
    ax1 = axes[i, 0]
    sns.histplot(data, kde=False, bins=30, color='skyblue', ax=ax1, stat="density")

    # Fit a normal distribution to the data and plot it
    mu, std = stats.norm.fit(data)
    xmin, xmax = ax1.get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)
    ax1.plot(x, p, 'k', linewidth=2)
    ax1.set_title(f'{column} - Normal Distribution')
    ax1.set_xlabel('Value')
    ax1.set_ylabel('Density')

    # Plot the box plot (on the second axis)
    ax2 = axes[i, 1]
    sns.boxplot(x=data, color='lightgreen', ax=ax2)
    ax2.set_title(f'{column} - Box Plot')
    ax2.set_xlabel('Value')

plt.tight_layout()
plt.show()
```

##Conclusion:
###As we can see above in the charts there are outliers in (Age and Weight) so the best action we can take to fill in missing values and the missing values in height is :
  * Age: Use Median (skewed distribution with outliers).
  * Height: Use Mean (approximately normal distribution).
  * Weight: Use Median (skewed distribution with many outliers).

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Height'] = df['Height'].fillna(df['Height'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].median())
```

```python
# Grouping the data by 'Sport' and calculating the mean for 'Weight', 'Height', and 'Age'
sport_group = df.groupby('Sport').agg({'Weight': 'mean', 'Height': 'mean', 'Age': 'mean'})

# Printing the result
print(sport_group)


```

```python


# Filter the DataFrame for the specific sports (Wrestling and Basketball)
selected_sports = ['Basketball', 'Wrestling']
filtered_df = df[df['Sport'].isin(selected_sports)]

# Group the filtered data by 'Sport' and calculate the mean for 'Weight', 'Height', and 'Age'
sport_group = filtered_df.groupby('Sport').agg({'Weight': 'mean', 'Height': 'mean', 'Age': 'mean'})

# Set up the plotting environment
plt.figure(figsize=(12, 6))

# Create custom color palette: Orange for Basketball, and default color for others
custom_palette = ['orange' if sport == 'Basketball' else 'skyblue' for sport in sport_group.index]

# Plotting Weight, Height, and Age for each selected Sport
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotting Weight
sns.barplot(x=sport_group.index, y=sport_group['Weight'], ax=axes[0], palette=custom_palette)
axes[0].set_title('Average Weight by Sport')
axes[0].set_xlabel('Sport')
axes[0].set_ylabel('Average Weight (kg)')

# Plotting Height
sns.barplot(x=sport_group.index, y=sport_group['Height'], ax=axes[1], palette=custom_palette)
axes[1].set_title('Average Height by Sport')
axes[1].set_xlabel('Sport')
axes[1].set_ylabel('Average Height (cm)')

# Plotting Age
sns.barplot(x=sport_group.index, y=sport_group['Age'], ax=axes[2], palette=custom_palette)
axes[2].set_title('Average Age by Sport')
axes[2].set_xlabel('Sport')
axes[2].set_ylabel('Average Age (years)')

# Adjust layout
plt.tight_layout()
plt.show()

```

##Conclusion:
* Basketball players are generally taller but slightly lighter compared to wrestlers. This is because basketball favors height for gameplay advantages like scoring and defending.

* Wrestlers, on the other hand, tend to be heavier, likely due to the sport's emphasis on strength and power.

* The average age of athletes in both sports is quite similar, indicating that these sports attract individuals of comparable age ranges.

* This comparison highlights how different sports influence the physical attributes of their athletes, tailored to the demands of each discipline.
