###Dealing with the categorical data and scaling numerical data
There is one categorical feature called (education) I will deal with a label encoding method because it Appears specific ordinal other categorical features with one hot encoding method.
As for the numerical features, I will use Standardization Scaling.

```python
df.education.value_counts()
```

###Using a map function (label encoding method) on education feature and Data Preprocessing with pipeline

```python
education_mapping = {
    'Preschool': 0,
    '1st-4th': 1,
    '5th-6th': 2,
    '7th-8th': 3,
    '9th': 4,
    '10th': 5,
    '11th': 6,
    '12th' :7,
    'HS-grad': 8,
    'Assoc-voc': 9,
    'Assoc-acdm': 10,
    'Some-college': 11,
    'Bachelors': 12,
    'Masters': 13,
    'Doctorate': 14,
    'Prof-school': 15
}
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

df = pd.DataFrame(df)

df['education_numeric'] = df['education'].map(education_mapping)

numeric_features = ['age', 'capital-gain', 'capital-loss', 'hours-per-week']
categorical_features = ['workclass', 'occupation', 'relationship', 'race', 'gender']

# Define transformers
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(sparse_output=False, drop='first')

# Create preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Drop original 'education' column
df = df.drop('education', axis=1)

# Apply preprocessing
processed_data = pipeline.fit_transform(df.drop(['education_numeric','income'],axis=1))

# Get column names for processed data
columns = numeric_features + list(pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(categorical_features))
processed_df = pd.DataFrame(processed_data, columns=columns)

# Add back 'education_numeric'
processed_df['education_numeric'] = df['education_numeric'].values

# Split data
X_train, X_test, y_train, y_test = train_test_split(processed_df, df['income'], test_size=0.2, random_state=42)
```

### We have unbalance data so I will use smote method to create more data to make the data more balance.

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                            weights=[0.1, 0.9], flip_y=0, random_state=42)


print("Before SMOTE")
print(pd.Series(y_train).value_counts())

smote = SMOTE(random_state=42)

X_res, y_res = smote.fit_resample(X_train, y_train)

print("\nAfter SMOTE")
print(pd.Series(y_res).value_counts())

```
