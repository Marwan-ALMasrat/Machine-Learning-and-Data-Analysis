```python
from sklearn.base import BaseEstimator, ClassifierMixin
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, StackingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.utils.validation import check_is_fitted

class XGBClassifierWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, **params):
        self.model = XGBClassifier(**params)

    def fit(self, X, y):
        self.model.fit(X, y)
        self.classes_ = self.model.classes_
        return self

    def predict(self, X):
        check_is_fitted(self)
        return self.model.predict(X)

    def predict_proba(self, X):
        check_is_fitted(self)
        return self.model.predict_proba(X)



rf_clf = RandomForestClassifier(random_state=42, n_estimators=300, max_depth=10)
xgb_clf = XGBClassifierWrapper(random_state=42, n_estimators=300, max_depth=100)
meta_model = LogisticRegression()

stacking_clf = StackingClassifier(
    estimators=[('rf', rf_clf), ('xgb', xgb_clf)],
    final_estimator=meta_model
)

stacking_clf.fit(X_train, y_train)

y_pred = stacking_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Classification Report:\n", classification_report(y_test, y_pred))
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
import numpy as np

# One-Hot Encoding Mappings (example mappings, adjust as needed)
one_hot_dict = {
    "sex": {
        "Male": [1, 0],
        "Female": [0, 1]
    },
    "dataset": {
        "Cleveland": [1, 0, 0],
        "Hungary": [0, 1, 0],
        "Long Beach": [0, 0, 1]
    },
    "cp": {
        "typical angina": [1, 0, 0, 0],
        "atypical angina": [0, 1, 0, 0],
        "non-anginal pain": [0, 0, 1, 0],
        "asymptomatic": [0, 0, 0, 1]
    },
    "fbs": {
        "True": [1],
        "False": [0]
    },
    "restecg": {
        "normal": [1, 0],
        "st-t wave abnormality": [0, 1],
        "lv hypertrophy": [0, 0, 1]
    },
    "slope": {
        "upsloping": [1, 0, 0],
        "flat": [0, 1, 0],
        "downsloping": [0, 0, 1]
    },
    "thal": {
        "normal": [1, 0, 0],
        "fixed defect": [0, 1, 0],
        "reversable defect": [0, 0, 1]
    },
    "exang": {
        "True": [1],
        "False": [0]
    }
}

# New input record
record = {
    "age": 63,  # Age
    "sex": "Male",
    "dataset": "Cleveland",
    "cp": "typical angina",
    "trestbps": 145.0,
    "chol": 233.0,
    "fbs": True,
    "restecg": "lv hypertrophy",
    "thalch": 150.0,
    "exang": False,
    "oldpeak": 2.3,
    "slope": "downsloping",
    "thal": "fixed defect",
    "num": 0
}

def encode_categorical(feature, value):
    return one_hot_dict[feature].get(value)

def preprocess_record(record):
    features = []

    features.extend(encode_categorical("sex", record["sex"]))  # Sex
    features.extend(encode_categorical("dataset", record["dataset"]))  # Dataset (Cleveland, Hungary, Long Beach)
    features.extend(encode_categorical("cp", record["cp"]))  # Chest pain type
    features.extend(encode_categorical("fbs", str(record["fbs"])))  # Fasting blood sugar
    features.extend(encode_categorical("restecg", record["restecg"]))  # Resting ECG
    features.extend(encode_categorical("slope", record["slope"]))  # Slope
    features.extend(encode_categorical("thal", record["thal"]))  # Thalassemia
    features.extend(encode_categorical("exang", str(record["exang"])))  # Exercise induced angina

    means = {
        "age": 52.9,
        "trestbps": 132.6,
        "chol": 245.9,
        "thalch": 141.02,
        "oldpeak": 0.867
    }
    stds = {
        "age": 9.5,
        "trestbps": 17.2,
        "chol": 57.5,
        "thalch": 24.1,
        "oldpeak": 1.063
    }

    features.append((record["age"] - means["age"]) / stds["age"])
    features.append((record["trestbps"] - means["trestbps"]) / stds["trestbps"])
    features.append((record["chol"] - means["chol"]) / stds["chol"])
    features.append((record["thalch"] - means["thalch"]) / stds["thalch"])
    features.append((record["oldpeak"] - means["oldpeak"]) / stds["oldpeak"])

    return np.array(features).reshape(1, -1)

preprocessed_input = preprocess_record(record)




print(preprocessed_input)

```
